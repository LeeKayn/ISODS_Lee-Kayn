```python
import cv2
import os
import shutil
from datetime import datetime

filename = '/content/ISODS_Lee-Kayn/Task2/Task2.mp4'
output_dir = 'output'

if not os.path.exists(output_dir):
    os.makedirs(output_dir)

cap = cv2.VideoCapture(filename)

if not cap.isOpened():
    print("Error: Could not open video file.")
    exit()

frame_count = 0
while cap.isOpened():
    ret, frame = cap.read()

    if ret:
        t = datetime.now()
        timestamp_str = t.strftime('%H%M%S%f')

        filename = os.path.join(output_dir, f"frame_{timestamp_str}.jpg")
        cv2.imwrite(filename, frame)

        frame_count += 1

        if frame_count % 100 == 0:
            print(f"Extracted and saved {frame_count} frames...")

    else:
        break

print(f"Successfully extracted {frame_count} frames from {filename} and saved them to {output_dir}")
```

```python
import glob
import json
import os
from PIL import Image
import torch
from transformers import (
    VisionEncoderDecoderModel,
    ViTFeatureExtractor,
    AutoTokenizer,
)

# Model and tokenizer loading (change path if needed)
model = VisionEncoderDecoderModel.from_pretrained(
    "nlpconnect/vit-gpt2-image-captioning"
)
feature_extractor = ViTFeatureExtractor.from_pretrained(
    "nlpconnect/vit-gpt2-image-captioning"
)
tokenizer = AutoTokenizer.from_pretrained("nlpconnect/vit-gpt2-image-captioning")

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model.to(device)

max_length = 32
num_beams = 8
gen_kwargs = {"max_length": max_length, "num_beams": num_beams}


def predict_step(image_path):
    i_image = Image.open(image_path)
    if i_image.mode != "RGB":
        i_image = i_image.convert(mode="RGB")

    pixel_values = feature_extractor(images=[i_image], return_tensors="pt").pixel_values
    pixel_values = pixel_values.to(device)

    output_ids = model.generate(pixel_values, **gen_kwargs)
    preds = tokenizer.decode(output_ids[0], skip_special_tokens=True)
    return preds.strip()


# Get image paths and sort them
image_paths = sorted(glob.glob("/content/ISODS_Lee-Kayn/Task2/*.jpg"))  # Modify pattern as needed

# Initialize variables for tracking predictions
current_prediction = None
current_prediction_files = []

# Create results dictionary
results = {}

# Predict and group descriptions
for image_path in image_paths:
    filename = os.path.basename(image_path)  # Extract filename with extension
    description = predict_step(image_path)

    # Check if the current prediction matches the previous one
    if description == current_prediction:
        current_prediction_files.append(filename)
    else:
        # Save the previous group of files if exists
        if current_prediction is not None:
            key = "-".join(current_prediction_files)
            # Only save unique filenames in the group
            unique_filenames = list(set(current_prediction_files))
            results[key] = current_prediction
        # Start a new group
        current_prediction = description
        current_prediction_files = [filename]

# Save the last group of files
if current_prediction is not None:
    key = "-".join(current_prediction_files)
    # Only save unique filenames in the group
    unique_filenames = list(set(current_prediction_files))
    results[key] = current_prediction

# Save results to JSON file
with open("descriptions_2.json", "w") as f:
    json.dump(results, f, indent=4)  # Ensure readability with indentation

print("Descriptions saved to descriptions.json")
```

```python
from IPython.core.display import HTML
import plotly.express as px
from PIL import Image
import numpy as np
import datetime
import cv2
import clip
import torch
import math

class video_parser:
    def __init__(self, file_path, search, batch_size=256):
        self.file_path = file_path
        self.search = search
        self.batch_size = batch_size

    def get_frames_from_file(self, N):
        video_frames = []

        capture = cv2.VideoCapture(self.file_path)
        fps = capture.get(cv2.CAP_PROP_FPS)

        current_frame = 0
        while capture.isOpened():
            ret, frame = capture.read()

            if ret == True:
                pil_frame = Image.fromarray(cv2.cvtColor(frame, cv2.COLOR_BGR2RGB))
                video_frames.append(pil_frame)
            else:
                break

            current_frame += N
            capture.set(cv2.CAP_PROP_POS_FRAMES, current_frame)

        print(f"Frames extracted: {len(video_frames)}")
        return video_frames, fps

    def text_analysis(self, search_query, model, device):
        with torch.no_grad():
            text_features = model.encode_text(clip.tokenize(search_query).to(device))
            text_features /= text_features.norm(dim=-1, keepdim=True)
        return text_features

    def get_clump(self, N, device, model, preprocess, search_query):
        video_frames, fps = self.get_frames_from_file(N)
        frames, fps = self.get_frames_from_file(15)

        print("Number of frames:", len(frames))

        groups = 10
        batches = len(frames) // groups

        video_frames_clump = []
        for j in range(groups):
            clump = preprocess(frames[j * batches])
            for i in range(1, batches):
                clump += preprocess(frames[i + (j * batches)])
            clump /= batches
            clump_deviation = clump * 0
            for s in range(batches):
                clump_deviation += clump - preprocess(frames[s + (j * batches)])
            video_frames_clump.append(clump_deviation)

        video_features = torch.empty([0, 512], dtype=torch.float16).to(device)

        for i in range(groups):
            batch_frames = video_frames_clump[i: i + 1]
            batch_preprocessed = torch.stack([frame for frame in batch_frames]).to(device)

            with torch.no_grad():
                batch_features = model.encode_image(batch_preprocessed)
                batch_features /= batch_features.norm(dim=-1, keepdim=True)

            video_features = torch.cat((video_features, batch_features))

        text_features = self.text_analysis(search_query, model, device)
        similarities = (100.0 * video_features @ text_features.T)
        v, bf = similarities.topk(1, dim=0)

        video_frames = frames[bf * batches:(bf + 1) * batches]
        return video_frames, fps

    def get_vid_feat(self, N, device, preprocess, model, batch_size, search_query):
        video_frames, fps = self.get_clump(N, device, model, preprocess, search_query)
        batch_size = batch_size
        batches = math.ceil(len(video_frames) / batch_size)

        video_features = torch.empty([0, 512], dtype=torch.float16).to(device)

        for i in range(batches):
            batch_frames = video_frames[i * batch_size: (i + 1) * batch_size]
            batch_preprocessed = torch.stack([preprocess(frame) for frame in batch_frames]).to(device)

            with torch.no_grad():
                batch_features = model.encode_image(batch_preprocessed)
                batch_features /= batch_features.norm(dim=-1, keepdim=True)

            video_features = torch.cat((video_features, batch_features))

        return video_features

    def search_video(self, search_query, N, device, preprocess, model, batch_size, display_heatmap=True, display_results_count=3):

        video_features = self.get_vid_feat(N, device, preprocess, model, batch_size, search_query)
        video_frames, fps = self.get_frames_from_file(N)

        text_features = self.text_analysis(search_query, model, device)

        similarities = (100.0 * video_features @ text_features.T)
        values, best_photo_idx = similarities.topk(display_results_count, dim=0)

        if display_heatmap:
            print("Search query heatmap over the frames of the video:")
            fig = px.imshow(similarities.T.cpu().numpy(), height=50, aspect='auto', color_continuous_scale='viridis')
            fig.update_layout(coloraxis_showscale=False)
            fig.update_xaxes(showticklabels=False)
            fig.update_yaxes(showticklabels=False)
            fig.update_layout(margin=dict(l=0, r=0, b=0, t=0))
            fig.show()
            print()

        for frame_id in best_photo_idx:
            display(video_frames[frame_id])


    def show_matches(self):
        device = "cuda" if torch.cuda.is_available() else "cpu"
        model, preprocess = clip.load("ViT-B/32", device=device)
        self.search_video(self.search, 64, device, preprocess, model, self.batch_size)

# Sử dụng ví dụ
file_path = "/content/Task3/Task3.mp4"
search_query = "group"
parser = video_parser(file_path, search_query)
parser.show_matches()
```
